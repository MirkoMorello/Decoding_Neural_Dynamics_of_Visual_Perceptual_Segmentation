```python
import numpy as np

import matplotlib.pyplot as plt

def plot_model_comparison(model_names, metrics, metric_values, title="Model Performance Comparison"):

    x = np.arange(len(metrics))  # the label locations
    width = 0.35  # the width of the bars
    
    fig, ax = plt.subplots(figsize=(12, 7))
    
    for i, (model_name, values) in enumerate(zip(model_names, metric_values)):
        offset = width * (i - 0.5 * (len(model_names) - 1))
        ax.bar(x + offset, values, width, label=model_name)
    
    ax.set_ylabel('Log Likelihood')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(metrics)
    ax.legend()
    
    # Add a grid for better readability
    ax.grid(True, linestyle='--', alpha=0.7)
    
    # Add values on top of bars
    for i, model_vals in enumerate(metric_values):
        offset = width * (i - 0.5 * (len(model_names) - 1))
        for j, val in enumerate(model_vals):
            ax.annotate(f'{val:.3f}', 
                       xy=(x[j] + offset, val), 
                       textcoords="offset points",
                       xytext=(0, 3),  # 3 points vertical offset
                       ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()

# Example usage
model_names = ["DeepGazeIII", "DinoGaze"]
metrics = ["Salicon", "Spatial Priority map", "Scanpath + Spatial Priority freezed", "Scanpath + Spatial Priority unfreezed"]
metric_values = [
    [0.7932, 2.0587, 2.510, 0],  # Model A performance
    [0.8159, 2.1474, 2.5576, 0]   # Model B performance
]

plot_model_comparison(model_names, metrics, metric_values)
```
